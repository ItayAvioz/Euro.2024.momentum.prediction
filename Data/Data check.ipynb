### Column: `possession`

- **Type:** Integer  
- **What it tells you:**  
  Uniquely identifies each “possession” in the game. A possession is a continuous period during which the ball is live and controlled by one team.  
- **Example:**  
  - `1` → the first possession of the match  
  - `23` → the 23rd time either team gained clear control of the ball  
- **Notes:**  
  - A new possession is logged whenever a team establishes clear control of the ball (e.g. after a turnover, blocked shot out‐of‐bounds, throw‐in).  
  - Even if the same team retains the ball through a block or rebound, it still counts as a new possession.

### Column: `play_pattern`

- **Type:** Categorical (integer codes mapped to play‐pattern names)  
- **What it tells you:**  
  Identifies the kind of passage of play leading into this event (e.g. set-piece, open play, counter-attack).

- **Codes & Meanings:**
  1. **Regular Play**  
     - The event was not part of any special play sequence.
  2. **From Corner**  
     - This event occurred during the attacking sequence immediately following a corner kick.
  3. **From Free Kick**  
     - This event occurred during the attacking sequence immediately following a free kick.
  4. **From Throw In**  
     - This event occurred during the attacking sequence immediately following a throw-in.
  5. **Other**  
     - The event was part of some other unusual play pattern not covered by the standard codes.
  6. **From Counter**  
     - The event was part of a fast break (counter-attack). In StatsBomb’s definition that means:
       - Possession began with an opposing turnover in your defensive or midfield third.
       - At least 75% of the possession’s progression (by chain metrics) was directed toward goal.
       - The counter-attack traveled at least 18 yards toward goal.
  7. **From Goal Kick**  
     - This event occurred during the sequence immediately following a goal kick.
  8. **From Keeper**  
     - This event occurred during the attacking sequence immediately following a goalkeeper distribution.
  9. **From Kick Off**  
     - This event occurred during the sequence immediately following kick-off.

- **Usage notes:**  
  - Use this to filter or group events by the type of play (e.g. isolating set-pieces, counter-attacks, open-play actions).  
  - The integer codes map to a small, fixed set of high-level play contexts.

````python
import pandas as pd
from collections import Counter

import ast

def count_objects_in_arrays(df, column_name):
    """
    Count the number of objects in each array for each row
    and provide a summary of occurrences
    """
    # Count objects in each array, handling string representations of arrays
    def safe_len(x):
        if x is None or pd.isna(x):
            return 0
        if isinstance(x, (list, tuple)):
            return len(x)
        if isinstance(x, str):
            # Try to parse string representation of array
            try:
                # Remove any extra whitespace and parse as literal
                parsed = ast.literal_eval(x.strip())
                if isinstance(parsed, (list, tuple)):
                    return len(parsed)
                else:
                    return 1  # Single value
            except:
                # If parsing fails, treat as single string
                return 1
        # For other types (like float), return 0
        return 0
    
    object_counts = df[column_name].apply(safe_len)
    
    # Get summary of counts
    summary = Counter(object_counts)
    
    # Print summary
    print("Summary of object counts:")
    print("-" * 30)
    for count, occurrences in sorted(summary.items()):
        print(f"{count} objects - {occurrences} occurrences")
    
    return object_counts, summary

object_counts, summary = count_objects_in_arrays(df_events, 'related_events')
````

### Column: `location`

- **Type:** Array of two integers `[x, y]`  
- **What it tells you:**  
  The on-pitch coordinates of the event, where  
  - `x` is the horizontal distance from the left touchline (0 at left, 120 at right)  
  - `y` is the vertical distance from the bottom goal line (0 at bottom, 80 at top)

- **Example from diagram:**  
  - `60,40` is the exact center of the pitch.  
  - `6,30` is 6 yards in from the left touchline and 30 yards up from the bottom goal line.  
  - `120,18` is 18 yards up on the right touchline.
  
- **Points to consider:**
- location has the opposite meaning for teams.
- location alternates in all halves including overtime

### Pass Object (extended)

A “Pass” event has the following key properties:

- **recipient** (object)  
  - **id** / **name** – The UUID and name of the player for whom the pass was intended.

- **length** (decimal)  
  - Distance of the pass in **yards**.

- **angle** (decimal)  
  - Direction of the pass in **radians**, measured clockwise from straight ahead (`0` = forward, `π/2` = right, `–π/2` = left, `π` = backward).

- **height** (object)  
  - **id** / **name** – Peak height of the ball:  
    1. `Ground Pass`  
    2. `Low Pass` (below shoulder height)  
    3. `High Pass` (above shoulder height)

- **end_location** (array [x, y])  
  - Pitch coordinates (0–120 by 0–80) where the ball landed.

- **cross** (boolean)  
  - `true` if this pass was a cross into the penalty area.

- **cut_back** (boolean)  
  - `true` if this pass was a low, backwards “cut-back” inside the opposition’s penalty box.

- **switch** (boolean)  
  - `true` if the pass moved play at least 50% of the pitch width (i.e. switched play to the opposite side).

- **shot_assist** (boolean)  
  - `true` if this pass led directly to a shot (that did not score).

- **goal_assist** (boolean)  
  - `true` if this pass directly created a goal.

- **body_part** (object)  
  - **id** / **name** – The body part used to make the pass, e.g.  
    - `40` / `Right Foot`  
    - `37` / `Head`  
    - `68` / `Drop Kick`  
    - `69` / `Keeper Arm`  
    - `70` / `Other`

- **type** (object)  
  - **id** / **name** – The attribute specifying the kind of pass, e.g.  
    - `61` / `Corner`  
    - `62` / `Free Kick`  
    - `63` / `Goal Kick`  
    - `64` / `Interception`  
    - `65` / `Throw‐in`

- **assisted_shot_id** (UUID or null)  
  - If this pass created a shot, the UUID of that shot event.

- **deflected** (boolean)  
  - `true` if the ball was deflected off another player en route.

- **miscommunication** (boolean)  
  - `true` if the pass was accurate but the recipient misread it.

---

### Dribble Object Fields

A **Dribble** event captures an individual one-on-one beat-attempt by a player.  Key subfields include:

| Field      | Type    | Description                                                                                     |
|:-----------|:--------|:-----------------------------------------------------------------------------------------------|
| **overrun**  | boolean | `TRUE` if the dribbler goes past the original defender into possession of another player.      |
| **nutmeg**   | boolean | `TRUE` if the dribble was played through an opponent’s legs.                                    |
| **outcome**  | object  | Result of the dribble attempt:  
  - **id 8 / “Complete”** → Successful dribble past the defender.  
  - **id 9 / “Incomplete”** → Failed to beat the defender; possession lost or held up.  
| **no_touch** | boolean | `TRUE` if the “dribble” was actually a pass around the opponent rather than a carry past them.  |

---

**Usage notes:**  
- Filter on `outcome.name == "Complete"` to isolate successful take-ons.  
- Combine `overrun` and `nutmeg` flags to find the most dominant dribbles.  
- Exclude rows with `no_touch == TRUE` if you only want pure carry-style dribbles.

| Variable         | Type       | Description                                                                                                                                                            |
|:-----------------|:-----------|:-----------------------------------------------------------------------------------------------------------------------------------------------------------------------|
| **event_uuid**   | UUID       | The unique identifier of the shot event this frame corresponds to.                                                                                                     |
| **visible_area** | Array      | A closed‐loop polygon of pitch coordinates `[X1, Y1, X2, Y2, …, Xn, Yn]` showing exactly what area was in the broadcast camera’s view when the frame was captured.    |
| **freeze_frame** | Array      | An array of player‐snapshot objects (see below). Unlike the main event data, these frames only label the shooting player; teammates and opponents appear generically.   |

#### Player‐Snapshot Object

Each entry in `freeze_frame` describes one player’s position and role at the moment of the shot:

| Field      | Type      | Description                                                                                                              |
|:-----------|:----------|:-------------------------------------------------------------------------------------------------------------------------|
| **location** | Array `[x, y]` | The player’s pitch coordinates, oriented so that their team’s attack is always 0→120 on the X axis.                       |
| **teammate** | boolean   | `TRUE` if the player is on the same team as the shooter.                                                                  |
| **actor**    | boolean   | `TRUE` for the shooter themself; all other players will have this set to `FALSE`.                                         |
| **keeper**   | boolean   | `TRUE` if this player is the goalkeeper (for either team).                                                                |
